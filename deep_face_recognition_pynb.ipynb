```python
from mtcnn.mtcnn import MTCNN
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import math
image_path = '/content/1.jpeg'
img = plt.imread(image_path)
plt.imshow(img)
plt.show()
```

```python
face_detector = MTCNN()
results = face_detector.detect_faces(img)
```

```python
# draw an image with detected objects
def count_faces(filename, result_list):
    # load the image
    data = plt.imread(filename)
    # plot the image
    plt.imshow(data)
    c = 0
    # get the context for drawing boxes
    ax = plt.gca()
    # plot each box
    for result in result_list:
        c += 1
        # get coordinates
        x, y, width, height = result['box']
        # create the shape
        rect = plt.Rectangle((x, y), width, height, fill=False, color='green')
        # draw the box
        ax.add_patch(rect)
        # show the plot
    plt.show()
    return c
```

```python
print("\nTotal no of faces in the image are : " ,count_faces(image_path, results))
```

```python
# draw an image with detected objects
def draw_facebox_and_keypoints(filename, result_list):
    # load the image
    data = plt.imread(filename)
    # plot the image
    plt.imshow(data)
    # get the context for drawing boxes
    ax = plt.gca()
    # plot each box
    for result in result_list:
        # get coordinates
        x, y, width, height = result['box']
        # create the shape
        rect = plt.Rectangle((x, y), width, height,fill=False, color='green')
        # draw the box
        ax.add_patch(rect)
        # draw the dots
        for key, value in result['keypoints'].items():
            # create and draw dot
            dot = plt.Circle(value, radius=1, color='red')
            ax.add_patch(dot)
            # show the plot
    plt.show()# filename = 'test1.jpg' # filename is defined above, otherwise uncomment
```

```python
draw_facebox_and_keypoints('/content/1.jpeg', results)
```
